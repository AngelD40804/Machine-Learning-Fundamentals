```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC , OneClassSVM
from sklearn.datasets import make_blobs
from numpy import quantile, where, random
import numpy as np

import matplotlib.pyplot as plt


iris = load_iris()
iris.feature_names
# print(iris.feature_names[])

print(iris.data[0:5,:2])
print(iris.target[0:5])

# print(iris.data)

x = iris.data[iris.target != 2,0:2]
# print(x)
y= iris.target[iris.target!=2]

x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2)
SVMmodel=SVC(kernel='linear')
SVMmodel.fit(x_train,y_train)
SVMmodel.get_params()
SVMmodel.score(x_test,y_test)
# print(y_test)
# print(x_test[:,1])
# plt.scatter(x_test[:,0], y_test)

supvec = SVMmodel.support_vectors_

W = SVMmodel.coef_
b = SVMmodel.intercept_

random.seed(11)
x, _ = make_blobs(n_samples=300, centers=1, cluster_std=.3, center_box=(4, 4))
plt.scatter(x[:,0], x[:,1])
plt.scatter(supvec[:, 0], supvec[:, 1])
xgr =  np.linspace(min(x[:,0]), max(x[:,0]),100)
# plt.plot(xgr)
plt.show()

SVMmodelOne = OneClassSVM(kernel='rbf', gamma=0.001, nu=0.03)

SVMmodelOne.fit(x)
pred = SVMmodelOne.predict(x)
print(pred)
anom_index = where(pred==-1)



# scores = SVMmodelOne.score_samples(x)

# thresh = quantile(scores, 0.01)
# print(thresh)
# index = where(scores<=thresh)
# values = x[index]

# plt.scatter(x[:,0], x[:,1])
# plt.scatter(values[:,0], values[:,1], color='red')
# plt.axis('equal')
# plt.show()

# print(x_train.shape)
# print(x_test.shape)
# x.shape

# print(y_test)
# print(y_train)
# supvectors = SVMmodel.support_vectors_
# W=SVMmodel.coef_
# b=SVMmodel.intercept_
# xgr=np.linspace(min(x[:,0]), max(x[:,0]),100) (with numpy)
```

```python
SVMmodelOne = OneClassSVM(kernel='rbf', gamma=0.01, nu=0.5)


SVMmodelOne.fit(x)
pred = SVMmodelOne.predict(x)
anom_index = where(pred==-1)
values = x[anom_index]
print(pred)
plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='red')
plt.axis('equal')
plt.show()
```

```python

scores = SVMmodelOne.score_samples(x)

thresh = quantile(scores, .75)
print(thresh)
index = where(scores<=thresh)
values = x[index]
print(scores)

plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0], values[:,1], color='red')
plt.axis('equal')
plt.show()

```
